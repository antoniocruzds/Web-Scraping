# Análise de Salarios pagos à Contratados da UFBA
Primeira análise feita por mim (sozinho), com o bjetivo de identificar quais contratados (nomes e profissões) tem a melhor remuneração do corpo de contatados pela Universidade Federal da Bahia

Dados -> 'http://www.cgp.ufba.br/transparencia.asp?letra=todos'

```python
# Importando as bibliotecas
import pandas as pd
import requests as rq
```

```python
#anexando o Url
url = 'http://www.cgp.ufba.br/transparencia.asp?letra=todos'

#Requisitando a pagina web, e guardando a resposta
resposta = rq.get(url)

#Exibir a resposta obtida em forma de texto
resposta.text
```

```python
# convertendo a resposta em HTML e mostrando a tabela obtida
table = pd.read_html(resposta.text)

table
```

```python
# A primeira linha n é tabela, mas está aparecendo ali, precisamos selecionar só o que queremos
# A coluna que queremos é a 2
# vamos fazer um df apenas para essa tabela

df = table[1] # 1 pq começa a contar de 0

# Vamos começar a analisar as informações recebidas
df.head(20)# Exibindo as 20 primeiras linhas
```

```python
#Pedindo pra ele exibir as 20 ultimas e fazendo uma verificação
df.tail(20)
```

Analisando o df obitido, podemos perceber que ele ainda necessita de melhoras. iremos colocar a linha 0 como colunas, eliminar a coluna 3, e a ultima linha

```python
# Colocando a linha 0 como coluna
df.columns = df.iloc[0]

# Excluindo a ultima coluna
df = df.drop(df.columns[[3]], axis=1)

# Excluido a linha 0
df.drop(df.index[0], inplace=True)


#Verificando:
# Lendo as primeiras colunas
df.head()
```

```python
# Excluido a ultima linha
df = df[:-1]

# Lendo as ultimas
df.tail()
```

```python
# Vamos analisar os tipos das colunas
df.info()
```

Aqui podemos perceber que a coluna de remuneração não é numeral, e precisaremos modificar ela. (tbm iremos renomear pra facilitar o acesso)

```python
#Renomeando as colunas
df.columns = ['Nome', 'Cargo', 'Remuneracao']
```

```python
# Retirando o 'R$' da coluna de remuneração
df['Remuneracao'] = df['Remuneracao'].apply(lambda x: str(x). replace('R$ ',''))
```

```python
#Verificando se ele saiu e observando o dtype
df['Remuneracao'].value_counts()
```

Precisaremos que a coluna seja de float, logo iremos converte-lá

```python
# Tentando transformar em float
df['Remuneracao'] = df['Remuneracao'].astype('float64')
```

Deu erro pq o pandas ainda está identificando o numero como str. teremos que temover o primeiro ponto do numero, e ficar só com o segundo, informando a parte decimal;

ex: 11.800,12 -> 11800.12

```python
# Removendo o primeiro ponto
df['Remuneracao'] = df['Remuneracao'].apply(lambda x: str(x). replace('.','', 1))
df['Remuneracao'] = df['Remuneracao'].apply(lambda x: str(x). replace(',','.', ))
```

```python
#Verificando:
# Lendo as primeiras colunas
df.head()
```

```python
# Verificando as informações do df
df.info()
```

a coluna ja está como num, mesmo não aparecendo alí, mas eu quero que ela seja um float64

```python
# Transformando em float64
df['Remuneracao'] = df['Remuneracao'].astype('float64')
```

```python
# Verificando
df.info()
```

# Dados arrumados, vamos começar a extrair informações deles

```python
# Verificando a quantidade de contratados por cargo
df['Cargo'].value_counts()
```

```python
# Qual o total de pagamentos feitos por cargo 
#agrupando por cargo com groupby, pegar os valores referentes à remuneração, e somar
df.groupby('Cargo').sum()
```

```python
# Excluindo a coluna de nomes (não me interessa analisar nomes agora)
df = df.drop(df.columns[[0]], axis=1)
```

```python
# Quais os 20 maiores valores de remuneração?
df.nlargest(20, 'Remuneracao')
```

```python
# Quais os 20 menores valores de remuneração?
df.nsmallest(20, 'Remuneracao')
```
